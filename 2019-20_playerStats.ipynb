```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import numpy.ma as ma
from scipy.stats import linregress
import json as json
from ohmysportsfeedspy import MySportsFeeds
from sports_api_key import api_key 

msf = MySportsFeeds(version="2.0")
msf.authenticate(api_key, 'MYSPORTSFEEDS')
```

```python
#output = msf.msf_get_data(league='nba',season='2018-2019-regular',feed='seasonal_games',format='json')
#print(output)

team_stats = ['teamStatsTotals']
team_data = []
stats_data = []

lastUpdate = ['lastUpdatedOn']
playerStats = ['playerStatsTotals']
#chedule = ['schedule']
for team in team_stats:
    response = msf.msf_get_data(league='nba',season='2019-2020-regular',feed='seasonal_player_stats',format='json')
    lastUpdate.append(response['lastUpdatedOn'])
    playerStats.append(response['playerStatsTotals'])
    #chedule.append(response['schedule'])
#print(response)
```

```python
playerStats[1]
playerDict = playerStats[1]
nbaPlayerStats = pd.DataFrame(playerDict)
nbaPlayerStats
#nbaData
```

```python
nbaPlayerName = pd.concat([nbaPlayerStats.player.apply(pd.Series)], axis=1)
nbaPlayerName.head()
```

```python
nbaPlayerTeam = pd.concat([nbaPlayerStats.team.apply(pd.Series)], axis=1)
nbaPlayerTeam.head()
```

```python
concPlayerStats = pd.concat([nbaPlayerStats.stats.apply(pd.Series)], axis=1)
concPlayerStats.head()
```

```python
PlayerStatsFG = pd.concat([concPlayerStats.fieldGoals.apply(pd.Series)], axis=1)
PlayerStatsFG.head()
```

```python
PlayerStatsOFF = pd.concat([concPlayerStats.offense.apply(pd.Series)], axis=1)
PlayerStatsOFF.head()
```

```python
firstName = nbaPlayerName['firstName']
lastName = nbaPlayerName['lastName']
playerPos = nbaPlayerName['primaryPosition']
playerTeam = nbaPlayerTeam['abbreviation']
playerfGpct = PlayerStatsFG['fgPct']
playerPTS = PlayerStatsOFF['pts']
playerPPG = PlayerStatsOFF['ptsPerGame']


filtPlayerStats = pd.DataFrame({'Last Name':lastName,
                              "First Name":firstName,
                              "Position":playerPos,
                              "Team":playerTeam,
                                "Points Per Game":playerPPG,
                              "Total FG %":playerfGpct,
                              "Total Points Scored":playerPTS,
})

```

```python
filtPlayerStats
```

```python
filtPlayerStats.sort_values(by=['Team','Total Points Scored'], inplace=True)
filtPlayerStats.head(20)
```

```python
filtPlayerStats.to_csv('output_data/2019-2020_player.csv')
```
