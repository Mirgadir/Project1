```python
import gmaps
import pandas as pd

from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create a dataframe with coordinate data
venue_df = pd.read_csv('Raw_Data/NBA_venua.csv')
venue_df.dropna(inplace=True)
venue_df.head()
```

```python
# Store the coordinates in locations
locations = venue_df[["Latitude","Longitude"]]
```

```python
# Customize the size of the figure
figure_layout = {}

fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
markers = gmaps.marker_layer(locations)

# Add the layer to the map
fig.add_layer(markers)
fig
```

```python
# Create a dataframe for winning pct by team
winpct_df = pd.read_csv('Raw_Data/2019-20_TEAMSTATS.csv')
winpct_df.dropna(inplace=True)
winpct_df.head()
```

```python
# Merge venue_df & winpct_df to create a dataframe for the heatmap
heatmap_df = pd.merge(venue_df, winpct_df, how='outer', on='Team Name')
heatmap_df.head()
```

```python
# Store latitude and longitude in heatmap_locations
heatmap_locations = heatmap_df[["Latitude","Longitude"]]

# Fill NaN values and convert to float
win_pct = heatmap_df["Win Pct"].astype(float)
```

```python
# Plot Heatmap
fig_1 = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(heatmap_locations, weights=win_pct, dissipating=False, max_intensity=1.5, point_radius=2.0)
# print(max(win_pct))
# Add layer
fig_1.add_layer(heat_layer)

# Display figure
fig_1
```
