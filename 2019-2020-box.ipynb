```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import numpy.ma as ma
from scipy.stats import linregress
import json as json
from ohmysportsfeedspy import MySportsFeeds
from sports_api_key import api_key 

msf = MySportsFeeds(version="2.0")
msf.authenticate(api_key, 'MYSPORTSFEEDS')
```

```python
#output = msf.msf_get_data(league='nba',season='2018-2019-regular',feed='seasonal_games',format='json')
#print(output)

team_stats = ['teamStatsTotals']
team_data = []
stats_data = []

lastUpdate = ['lastUpdatedOn']
games = ['games']
#chedule = ['schedule']
for team in team_stats:
    response = msf.msf_get_data(league='nba',season='2019-2020-regular',feed='seasonal_games',format='json')
    lastUpdate.append(response['lastUpdatedOn'])
    games.append(response['games'])
    #chedule.append(response['schedule'])
print(response)
```

```python
nbaDict = games[1]
nbaData = pd.DataFrame(nbaDict)
nbaData
```

```python
nbaDataSchedule = pd.concat([nbaData.schedule.apply(pd.Series)], axis=1)
```

```python
nbaDataSchedule.head()
```

```python
nbaDataScore = pd.concat([nbaData.score.apply(pd.Series)], axis=1)
nbaDataScore.head()
```

```python
nbaData_road = pd.concat([nbaDataSchedule.awayTeam.apply(pd.Series)], axis=1)
nbaData_road.head(10)
```

```python
nbaData_home = pd.concat([nbaDataSchedule.homeTeam.apply(pd.Series)], axis=1)
nbaData_home.head(10)
```

```python
nbaData_venue = pd.concat([nbaDataSchedule.venue.apply(pd.Series)], axis=1)
nbaData_venue.head(10)
```

```python
nbaRoadAbr = nbaData_road['abbreviation']
nbaHomeAbr = nbaData_home['abbreviation']
nbaVen = nbaData_venue['name']
nbascoresRoad = nbaDataScore['awayScoreTotal']
nbascoresHome = nbaDataScore['homeScoreTotal']
nbaDate = nbaDataSchedule['startTime']


filteredNBAdata = pd.DataFrame({"Date":nbaDate,
                                "Home Team":nbaHomeAbr,
                                "- Score":nbascoresHome,
                                "Road Team":nbaRoadAbr,
                                " - Score":nbascoresRoad,
                                "Venue":nbaVen,})
```

```python
filteredNBAdata.to_csv('output_data/2019-2020-NBA.csv')
filteredNBAdata.head()
```

```python
#team_stats = ['teamStatsTotals']
team_data = []
stats_data = []

lastUpdate = ['lastUpdatedOn']
references = ['references']
#chedule = ['schedule']
for team in team_stats:
    response = msf.msf_get_data(league='nba',season='2019-2020-regular',feed='seasonal_games',format='json')
    lastUpdate.append(response['lastUpdatedOn'])
    references.append(response['references'])
    #chedule.append(response['schedule'])
#print(response)
```

```python
references
```

```python
nbaRefDict = references[1]
nbaRefData = pd.DataFrame.from_dict(nbaRefDict, orient='index')
nbaRefData
```

```python
nbaRef = nbaRefData.transpose()
nbaRef.head()
```

```python
nbaTeamRef = pd.concat([nbaRef.teamReferences.apply(pd.Series)], axis=1)
nbaTeamRef.head()
```

```python
nbaVenRef = pd.concat([nbaRef.venueReferences.apply(pd.Series)], axis=1)
nbaVenRef.head()
```

```python
nbaGeoRef = pd.concat([nbaVenRef.geoCoordinates.apply(pd.Series)], axis=1)
nbaGeoRef.head()
```

```python
teamName = nbaTeamRef['name']
teamAbr = nbaTeamRef['abbreviation']
teamCity = nbaVenRef['city']
teamArena = nbaVenRef['name']
teamLat = nbaGeoRef['latitude']
teamLng = nbaGeoRef['longitude']

filteredNBAref = pd.DataFrame({"Team Name":teamName,
                                "Abbreviation":teamAbr,
                                "City, State": teamCity,
                                "Venue":teamArena,
                                "Latitude":teamLat,
                               "Longitude":teamLng})
```

```python
filteredNBAref.to_csv('output_data/NBA_venua.csv')
filteredNBAref
```
